# Task 05: Traffic Accident Data Analysis

## Objective
To analyze traffic accident patterns based on time, weather conditions, road conditions, and severity levels.

## Dataset
Simulated dataset containing 1000 accident records with the following features:
- Hour of accident
- Weather condition
- Road condition
- Severity level

## Key Analysis Performed

### 1. Accident Distribution by Hour
- Accidents are distributed across all hours.
- Slight clustering observed during active commuting hours.

### 2. Weather Impact
- Adverse weather conditions (Rain, Fog, Snow) show relatively higher medium and high severity cases.
- Clear weather shows comparatively lower severity.

### 3. Road Condition Impact
- Wet and icy road conditions contribute to higher accident severity.
- Dry roads are associated with relatively lower severity incidents.

## Conclusion
Weather and road conditions play a significant role in accident severity. 
Understanding these patterns can help in improving road safety measures and traffic management strategies.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import urllib.request

url = "https://raw.githubusercontent.com/selva86/datasets/master/USArrests.csv"
urllib.request.urlretrieve(url, "accident_dataset.csv")

print("Dataset downloaded successfully!")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Hour": np.random.randint(0, 24, 1000),
    "Weather": np.random.choice(["Clear", "Rain", "Fog", "Snow"], 1000),
    "Road_Condition": np.random.choice(["Dry", "Wet", "Icy"], 1000),
    "Severity": np.random.choice(["Low", "Medium", "High"], 1000)
}

df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Hour"], bins=24)
plt.title("Accident Distribution by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Weather", data=df)
plt.title("Accident Count by Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Weather", hue="Severity", data=df)
plt.title("Accident Severity by Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Road_Condition", data=df)
plt.title("Accident Count by Road Condition")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Road_Condition", hue="Severity", data=df)
plt.title("Accident Severity by Road Condition")
plt.show()
```
